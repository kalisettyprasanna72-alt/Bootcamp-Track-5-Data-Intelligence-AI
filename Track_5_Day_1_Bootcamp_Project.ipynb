Assignment: The Satellite Weather Image Grid Fixer
• Task: Imagine a satellite sends weather data as a messy 24 \times 24 grid of numbers. Write a program to reshape it into a 3D matrix representing morning, afternoon, and night blocks. Use a Boolean mask to find all temperature spots that are dangerously high, and replace those bad readings with the average temperature of the entire grid in one step without using any loops.


```python
# Import NumPy library for numerical operations
import numpy as np
```

```python
# Create a 24x24 grid with random temperature values
# Values range from 10°C to 49°C
data = np.random.randint(10, 50, size=(24, 24))

print(data)
```

```python
# Reshape the 2D grid into a 3D array
# Morning, Afternoon, Night
data_3d = data.reshape(3, 8, 24)

print(data_3d.shape)
```

```python
# Set dangerous temperature threshold
threshold = 40

# Create Boolean mask
mask = data_3d > threshold

print(mask)
```

```python
# Set dangerous temperature threshold
threshold = 40

# Create Boolean mask
mask = data_3d > threshold

print(mask)
```

```python
# Calculate global average temperature
avg_temp = data_3d.mean()

# Replace dangerous temperatures with average
data_3d[mask] = avg_temp

print("Processed Data:\n", data_3d)
print("\nAverage Temperature:", avg_temp)
```
