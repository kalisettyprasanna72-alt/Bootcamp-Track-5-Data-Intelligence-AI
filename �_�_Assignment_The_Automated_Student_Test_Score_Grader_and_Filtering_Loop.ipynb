```python
# Student records using Dictionary

students = {
    "Rahul": 95,
    "Anjali": 82,
    "Kiran": 67,
    "Sneha": 45,
    "Arjun": 30,
    "Priya": 76,
    "Ravi": 55
}
```

```python
total = 0
highest = -1
lowest = 101
```

```python
print("STUDENT REPORT")
print("-" * 40)
```

```python
for name, marks in students.items():
  total += marks
```

```python
if marks > highest:
    highest = marks

if marks < lowest:
    lowest = marks
```

```python
if marks >= 90:
    grade = "A+"

elif marks >= 80:
    grade = "A"

elif marks >= 70:
    grade = "B"

elif marks >= 60:
    grade = "C"

elif marks >= 35:
    grade = "Pass"

else:
    grade = "Fail"
```

```python
status = "Pass" if marks >= 35 else "Fail"
```

```python
print(f"Name : {name}")
print(f"Marks : {marks}")
print(f"Grade : {grade}")
print(f"Status : {status}")
print("-" * 40)
```

```python
average = total / len(students)
```

```python
print("\nClass Summary")
print("Average Marks :", average)
print("Highest Marks :", highest)
print("Lowest Marks :", lowest)
```

```python
print("\nStudents Scoring Above 75")

for name, marks in students.items():

    if marks > 75:
        print(name, "-", marks)
```
