```python
# Cell 1: Data Storage & Global Setup

students = []  # Each student is a dict: {roll, name, branch, marks: {subject: mark}, average, grade}

SUBJECTS = ["Math", "Physics", "Chemistry", "English", "Computer Science"]

def find_student_index(roll_no=None, name=None):
    """Returns index of student in list matching roll_no or name, else -1"""
    for i, s in enumerate(students):
        if roll_no is not None and s["roll"] == roll_no:
            return i
        if name is not None and s["name"].lower() == name.lower():
            return i
    return -1

print("Data storage initialized. Subjects:", SUBJECTS)
```

```python
# Cell 2: Add New Student

def add_student():
    print("\n--- Add New Student ---")
    roll = input("Enter Roll Number: ").strip()

    if find_student_index(roll_no=roll) != -1:
        print("A student with this Roll Number already exists!")
        return

    name = input("Enter Name: ").strip()
    branch = input("Enter Branch: ").strip()

    marks = {}
    for subject in SUBJECTS:
        while True:
            try:
                mark = float(input(f"Enter marks in {subject} (out of 100): "))
                if 0 <= mark <= 100:
                    marks[subject] = mark
                    break
                else:
                    print("Marks must be between 0 and 100.")
            except ValueError:
                print("Invalid input. Please enter a number.")

    student = {
        "roll": roll,
        "name": name,
        "branch": branch,
        "marks": marks,
        "average": None,
        "grade": None
    }

    students.append(student)
    print(f"Student '{name}' added successfully!")
```

```python
# Cell 3: View All Students

def view_all_students():
    print("\n--- All Student Records ---")
    if not students:
        print("No student records found.")
        return

    for s in students:
        print("-" * 50)
        print(f"Roll No   : {s['roll']}")
        print(f"Name      : {s['name']}")
        print(f"Branch    : {s['branch']}")
        print("Marks     :", ", ".join(f"{sub}: {m}" for sub, m in s['marks'].items()))
        avg = s['average'] if s['average'] is not None else "Not calculated"
        grade = s['grade'] if s['grade'] is not None else "Not assigned"
        print(f"Average   : {avg}")
        print(f"Grade     : {grade}")
    print("-" * 50)
```

```python
# Cell 4: Search, Update, Delete Student

def search_student():
    print("\n--- Search Student ---")
    choice = input("Search by (1) Roll Number or (2) Name: ").strip()

    if choice == "1":
        roll = input("Enter Roll Number: ").strip()
        idx = find_student_index(roll_no=roll)
    elif choice == "2":
        name = input("Enter Name: ").strip()
        idx = find_student_index(name=name)
    else:
        print("Invalid choice.")
        return

    if idx == -1:
        print("Student not found.")
    else:
        s = students[idx]
        print("\nStudent Found:")
        print(f"Roll No: {s['roll']}, Name: {s['name']}, Branch: {s['branch']}")
        print("Marks:", s['marks'])
        print(f"Average: {s['average']}, Grade: {s['grade']}")


def update_student():
    print("\n--- Update Student Record ---")
    roll = input("Enter Roll Number of student to update: ").strip()
    idx = find_student_index(roll_no=roll)

    if idx == -1:
        print("Student not found.")
        return

    s = students[idx]
    print(f"Updating record for {s['name']} (leave blank to keep current value)")

    new_name = input(f"Name [{s['name']}]: ").strip()
    if new_name:
        s['name'] = new_name

    new_branch = input(f"Branch [{s['branch']}]: ").strip()
    if new_branch:
        s['branch'] = new_branch

    update_marks = input("Update marks? (y/n): ").strip().lower()
    if update_marks == "y":
        for subject in SUBJECTS:
            while True:
                val = input(f"{subject} [{s['marks'][subject]}]: ").strip()
                if val == "":
                    break
                try:
                    mark = float(val)
                    if 0 <= mark <= 100:
                        s['marks'][subject] = mark
                        break
                    else:
                        print("Marks must be between 0 and 100.")
                except ValueError:
                    print("Invalid input.")

    # Reset average/grade since marks may have changed
    s['average'] = None
    s['grade'] = None
    print("Student record updated successfully!")


def delete_student():
    print("\n--- Delete Student ---")
    roll = input("Enter Roll Number of student to delete: ").strip()
    idx = find_student_index(roll_no=roll)

    if idx == -1:
        print("Student not found.")
        return

    confirm = input(f"Are you sure you want to delete {students[idx]['name']}? (y/n): ").strip().lower()
    if confirm == "y":
        students.pop(idx)
        print("Student record deleted successfully!")
    else:
        print("Deletion cancelled.")
```

```python
# Cell 5: Class Average & Grade Assignment

def assign_grade(average):
    if average >= 75:
        return "Distinction"
    elif average >= 60:
        return "First Class"
    elif average >= 40:
        return "Second Class"
    else:
        return "Fail"


def calculate_student_average(student):
    marks = student['marks'].values()
    return sum(marks) / len(marks) if marks else 0


def calculate_class_average():
    print("\n--- Class Average ---")
    if not students:
        print("No student records found.")
        return

    # Ensure each student's individual average is calculated first
    total = 0
    for s in students:
        s['average'] = round(calculate_student_average(s), 2)
        total += s['average']

    class_avg = round(total / len(students), 2)
    print(f"Class Average (based on {len(students)} students): {class_avg}")


def assign_grades_to_all():
    print("\n--- Assign Grades to All Students ---")
    if not students:
        print("No student records found.")
        return

    for s in students:
        s['average'] = round(calculate_student_average(s), 2)
        s['grade'] = assign_grade(s['average'])
        print(f"{s['name']} (Roll {s['roll']}): Average = {s['average']}, Grade = {s['grade']}")

    print("Grades assigned to all students successfully!")
```

```python
# Cell 6: Filter by Grade & Find Class Topper

def filter_students_by_grade():
    print("\n--- Filter Students by Grade ---")
    if not students:
        print("No student records found.")
        return

    # Make sure grades are assigned
    for s in students:
        if s['grade'] is None:
            s['average'] = round(calculate_student_average(s), 2)
            s['grade'] = assign_grade(s['average'])

    print("Available Categories: Distinction, First Class, Second Class, Fail")
    category = input("Enter grade category to filter: ").strip().title()

    valid_categories = ["Distinction", "First Class", "Second Class", "Fail"]
    if category not in valid_categories:
        print("Invalid category.")
        return

    filtered = [s for s in students if s['grade'] == category]

    if not filtered:
        print(f"No students found in '{category}' category.")
        return

    print(f"\nStudents in '{category}' category:")
    for s in filtered:
        print(f"Roll {s['roll']} - {s['name']} - Average: {s['average']}")


def find_class_topper():
    print("\n--- Class Topper ---")
    if not students:
        print("No student records found.")
        return

    for s in students:
        if s['average'] is None:
            s['average'] = round(calculate_student_average(s), 2)

    max_avg = max(s['average'] for s in students)
    toppers = [s for s in students if s['average'] == max_avg]

    print(f"Highest Average: {max_avg}")
    print("Topper(s):")
    for s in toppers:
        print(f"Roll {s['roll']} - {s['name']} - Branch: {s['branch']}")
```

```python
# Cell 7: Menu-Driven Main Program

def main():
    while True:
        print("\n===== Student Performance Management System =====")
        print("1. Add New Student")
        print("2. View All Students")
        print("3. Search Student")
        print("4. Update Student Record")
        print("5. Delete Student")
        print("6. Calculate Class Average")
        print("7. Assign Grades to All Students")
        print("8. Filter Students by Grade")
        print("9. Find Class Topper")
        print("10. Exit")

        choice = input("Enter your choice: ").strip()

        if choice == "1":
            add_student()
        elif choice == "2":
            view_all_students()
        elif choice == "3":
            search_student()
        elif choice == "4":
            update_student()
        elif choice == "5":
            delete_student()
        elif choice == "6":
            calculate_class_average()
        elif choice == "7":
            assign_grades_to_all()
        elif choice == "8":
            filter_students_by_grade()
        elif choice == "9":
            find_class_topper()
        elif choice == "10":
            print("Exiting the system. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 10.")

# Run the program
main()
```
